```python
from qiskit import IBMQ, Aer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, transpile, assemble
from qiskit.tools.monitor import job_monitor
import matplotlib as mpl
# import basic plot tools
from qiskit.visualization import plot_histogram
```

```python
deutsch_circuit = QuantumCircuit(2,1)
deutsch_circuit.clear()
```

 Deutsch problem: Given an unknown unary function f, determine whether it is
balanced or constant in one querry of the corresponding quantum oracle Uf .


```python
def inputs_state(deutsch_circuit,a,b):
    # Circuit
    deutsch_circuit.barrier()
    deutsch_circuit.h([a,b])
    deutsch_circuit.barrier()

    #oracle

    deutsch_circuit.cx(a,b)
    deutsch_circuit.barrier()

    deutsch_circuit.h([a,b])
    deutsch_circuit.measure(a,0)
```

# Constant Function 

# input states = (0,0)

```python
inputs_state(deutsch_circuit,0,1)
deutsch_circuit.draw('mpl')
```

```python
aer_sim = Aer.get_backend('aer_simulator')   #Local Simulator
shots = 1024                                 #No. of times the circuit is running
qobj = assemble(deutsch_circuit, shots = shots) 
results = aer_sim.run(qobj).result()
counts = results.get_counts()
plot_histogram(counts) 
```

# input states = (1,1)

```python
deutsch_circuit.clear()
deutsch_circuit.x([0,1])
inputs_state(deutsch_circuit,0,1)
deutsch_circuit.draw('mpl')
```

```python
aer_sim = Aer.get_backend('aer_simulator')   #Local Simulator
shots = 1024                                 #No. of times the circuit is running
qobj = assemble(deutsch_circuit, shots = shots) 
results = aer_sim.run(qobj).result()
counts = results.get_counts()
plot_histogram(counts) 
```
